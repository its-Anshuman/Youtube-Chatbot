```python
from youtube_transcript_api import YouTubeTranscriptApi, CouldNotRetrieveTranscript
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import PromptTemplate
```

```python
!pip install python-certifi-win32
```

```python
video_id = "oX7OduG1YmI" # only the ID, not full URL
ytt_api = YouTubeTranscriptApi()

try:
    transcript_list = ytt_api.fetch(video_id, languages=['en'])
    transcript = " ".join(snippet.text for snippet in transcript_list)
    print(transcript)  # first 500 characters

except CouldNotRetrieveTranscript:
    print("Transcript not available for this video.")
    
```

```python
transcript_list
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = splitter.create_documents([transcript])
```

```python
len(chunks)
```

```python
chunks[100]
```

### Step 1c & 1d - Indexing (Embedding Generation and Storing in Vector Store)

```python
embedding = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLm-L6-v2')
embeddings =  embedding
vector_store = FAISS.from_documents(chunks, embeddings)
```

```python
vector_store.index_to_docstore_id
```

```python
vector_store.get_by_ids(['bbfe1e4f-ea29-4385-a3ca-62b5bb5ce600'])
```

### Step 2 - Retrieval


```python
retriever = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 4})
```

```python
retriever
```

```python
retriever.invoke('What is deepmind')
```

```python
llm = HuggingFaceEndpoint(repo_id="mistralai/Mistral-7B-Instruct-v0.2",task="text-generation")

model = ChatHuggingFace(llm=llm)
```

```python
prompt = PromptTemplate(
    template="""
      You are a helpful assistant.
      Answer ONLY from the provided transcript context.
      If the context is insufficient, just say you don't know.

      {context}
      Question: {question}
    """,
    input_variables = ['context', 'question']
)
```

```python
question          = "is the topic of nuclear fusion discussed in this video? if yes then what was discussed"
retrieved_docs    = retriever.invoke(question)
```

```python
retrieved_docs
```

```python
context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
context_text
```

```python
final_prompt = prompt.invoke({"context": context_text, "question": question})
```

```python
final_prompt
```

#### Step 4 - Generation

```python
answer = model.invoke(final_prompt)
print(answer.content)
```

### Building a Chain

```python
from langchain_core.runnables import RunnableParallel, RunnablePassthrough, RunnableLambda
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_docs(retrieved_docs):
    context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
    return context_text
```

```python
parallel_chain = RunnableParallel({
    'context': retriever | RunnableLambda(format_docs),
    'question': RunnablePassthrough()
})
```

```python
parallel_chain.invoke('Who is Demis?')
```

```python
parser = StrOutputParser()
```

```python
main_chain = parallel_chain | prompt | model | parser
```

```python
main_chain.invoke('Can You Summarize the Video?')
```
